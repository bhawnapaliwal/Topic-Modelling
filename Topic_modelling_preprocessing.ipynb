```python
index=-1
for line in lines:
#     if index==1:
#         break
    if len(line)!=0:
        if len(line)>=3 and line[0]=='U' and line[1]=='R' and line[2]=='L':
            parts = line.split('/')
            date=parts[3]+'/'+parts[4]+'/'+parts[5]
            category=parts[6]
            sub_category=''
            title=''
            content=''
            if len(parts)>7:
                if '.html' in parts[7]:
                    title = parts[7].replace('-', ' ')
                    title = title[:-5]
                else:
                    sub_category=parts[7]
                if len(parts)>8:
                    title = parts[8].replace('-', ' ')
                    title = title[:-5]
            print('date: '+date)
            print('category: '+category)
            print('sub-category: '+sub_category)
            print('title: '+title)
            print('content: '+content)
            df = df.append([{'Date':date,'Category':category,'Sub-Category':sub_category,'Title':title,'Content':content}])
#             print('hi df')
#             print(df)
            index=index+1
        else:
            if index>=0 and line!='\n':
                print('index: '+str(index))
                print(df.iloc[index])
                print('hello')
#                 print(df)
                df.iloc[index]['Content']=df.iloc[index]['Content']+line
```

```python
f = open('stop_words.txt','r')
lines = f.readlines()
f.close()
```

```python
ps = PorterStemmer() 
cnt=0
for i in range(8888):
    x = df.iloc[i]['Content']
    x_w = re.findall(r"[a-zA-Z]+",x)
    for i1 in range(len(x_w)):
        x_w[i1] = x_w[i1].lower()
        x_w[i1] = ps.stem(x_w[i1])
        if x_w[i1] in stopwords:
            x_w[i1]=''
        else:
            cnt=cnt+1
    s=''
    cnt = cnt+len(x_w)
    for i1 in range(len(x_w)):
        if x_w[i1]!='':
            s=s+x_w[i1]+' '
    df.iloc[i]['Content']=s
#     if i==0:
#         print(df.iloc[i]['Content'])
    if i%10==0:
        print("i: "+str(i))
print('cnt: '+str(cnt))
```
